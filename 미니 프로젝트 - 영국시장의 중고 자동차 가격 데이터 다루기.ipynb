```python
# 판다스 기능 사용
import pandas as pd

# 수치 및 연산 기능 사용
import numpy as np

# 통계 그래프 고급 시각화
import seaborn as sns

# 일반 그래프 시각화
import matplotlib.pyplot as plt
```

```python
# 파일불러오기 및 저장

cars_df = pd.read_csv('../data/cars.csv')
brand_df = pd.read_csv('../data/brand.csv')
```

```python
# 첫 5행 조회

cars_df.head()
```

```python
# 첫 5행 조회

brand_df.head()
```

```python
# cars_df 파일의 title 값 조회

cars_df['title'].unique()
```

```python
# brand_df 파일의 title 값 조회

brand_df['title'].unique()
```

```python
# title파일을 병합하기 위해 문자열을 맞춰준다.
# 브랜드명 차량명임을 확인했으니 띄어쓰기 기준 왼쪽문자+소문자만 필요로 함
# brand라는 컬럼명을 추가해서 값을 가져온다

cars_df['brand'] = cars_df['title'].str.split().str[0].str.lower()
```

```python
cars_df
```

```python
# cars_df를 왼쪽에, 왼쪽 파일의 brand열 기준, 오른쪽 파일의 title 기준 병합

cars_df = cars_df.merge(brand_df, how='left', left_on='brand', right_on='title')
```

```python
# 중복되는 컬럼 제거

cars_df = cars_df.drop(columns=['title_x', 'title_y'])
```

```python
cars_df
```

```python
cars_df.info()
```

```python
# L을 제거하고, float으로 변환

cars_df['Engine'] = cars_df['Engine'].str.strip('L').astype(float)
```

```python
# 띄어쓰기 기준으로 나눠서 두번째 문자만 가져온 후, float으로 변환
# int는 nan 값이 있으면 안됨
# 결측치 제거 후, int로 변환해 줌

cars_df['Emission Class'] = cars_df['Emission Class'].str.split().str[1].astype(float)
```

```python
cars_df.info()
```

```python
# 서비스 히스토리의 결측치 개수 확인

cars_df['Service history'].isna().sum()
```

```python
# 서비스 히스토리의 고유값 확인

cars_df['Service history'].unique()
```

```python
# 서비스 히스토리의 결측치를 Unknown으로 대체 (fillna)

cars_df['Service history'] = cars_df['Service history'].fillna('Unknown')
```

```python
# 서비스 히스토리 결측치 재확인

cars_df['Service history'].isna().sum()
```

```python
# 컬럼별 남은 결측치 개수 확인

cars_df.isna().sum()
```

```python
# 결측치 행 전부 삭제

cars_df = cars_df.dropna()
```

```python
# 결측치 재확인

cars_df.isna().sum()
```

```python
# 국가별 총 브랜드 개수
# groupby로 country를 묶어서 국가별 데이터를 합치고, 브랜드 열을 불러와 unique로 고유값의 수를 구함

# sort_values(ascending=False)

count = cars_df.groupby('country')['brand'].nunique()

count.index.name = None
print(f'국가별 총 브랜드개수\n\n{count}')
```

```python
# 가격, 주행거리, 연식, 이전 소유자, 엔진 배기량, 문 개수, 좌석수, 배기가스 등급
```

```python
cars_df['Price'].corr(cars_df['Mileage(miles)'])
```

```python
# 가격 기준으로 주행거리, 연식, 이전 소유자, 엔진 배기량, 문 개수, 좌석수, 배기가스 등급의 상관관계 값을 조회한다.

cars_df[
    ['Price', 'Mileage(miles)', 'Registration_Year', 'Previous Owners', 'Engine', 'Doors', 'Seats', 'Emission Class']
].corr()['Price'].drop('Price').sort_values(ascending=False)
```

```python
# 변수로 깔끔한 코드

car_corr = ['Price', 'Mileage(miles)', 'Registration_Year', 'Previous Owners', 'Engine', 'Doors', 'Seats', 'Emission Class']

cars_df[car_corr].corr()['Price'].drop('Price').sort_values(ascending=False)
```

```python
# 산점도 그래프로 이상치 확인하기

sns.scatterplot(x = cars_df.index, y = cars_df['Mileage(miles)'])
```

```python
# 내림차순으로 수치 확인하기

cars_df['Mileage(miles)'].sort_values(ascending = False)
```
